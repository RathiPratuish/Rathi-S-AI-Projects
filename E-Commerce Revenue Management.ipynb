####     The company charges

## Q 5. Find the net revenue generated by the E-Commerce company over all orders placed. (10 marks)

### Ans 5

```python
ECom_Data.sample()
```

```python
percent25 = ECom_Data[ECom_Data.discounted_price > 600]['discounted_price'].sum() * 0.25
percent15 = ECom_Data[(ECom_Data.discounted_price > 350) & (ECom_Data.discounted_price <= 600)]['discounted_price'].sum() * 0.15
percent10 = ECom_Data[(ECom_Data.discounted_price > 100) & (ECom_Data.discounted_price <= 350)]['discounted_price'].sum() * 0.10
others = ECom_Data[ECom_Data.discounted_price <= 100]['discounted_price'].sum() * 0.05
total = percent25 + percent15 + percent10 + others
print("Revenue of the company: \n25% on the orders having final price (discounted price) greater than 600: {:.2f}".format(percent25))
print("15% on the orders having final price (discounted price) greater than 350 but less than or equal to 600: {:.2f}".format(percent15))
print("10% on the orders having final price (discounted price) greater than 100 but less than or equal to 350: {:.2f}".format(percent10))
print("Otherwise, 5% on the final price (discounted price): {:.2f}".format(others))
print("Total Revenue of the company: {:.2f}".format(total))
```

```python
newIndexList = []
for i in ECom_Data.Order_Date:
    newIndexList.append(i.split('-')[1])
newIndexList # 01-12-2020 - [01, 12, 2020] - 12
```

```python
plt.figure(figsize=(6, 7), dpi = 100)
plt.title('Net Brand Revenue for South region')
plt.xlabel('Month of sales')
plt.ylabel('Sum of discounted price for South region (in 50k unit)')
sns.lineplot(net_rev_reg_month, x = net_rev_reg_month.index, y = net_rev_reg_month.columns[2], marker = 'o')
```

```python
plt.figure(figsize=(6, 7), dpi = 100)
plt.title('Net Brand Revenue for each region')
plt.xlabel('Month of sales')
plt.ylabel('Sum of discounted price for each region (in 1 million unit)')
sns.lineplot(net_rev_reg_month, x = net_rev_reg_month.index, y = net_rev_reg_month.columns[0], marker = 'o', label='East')
sns.lineplot(net_rev_reg_month, x = net_rev_reg_month.index, y = net_rev_reg_month.columns[1], marker = 'o', label='North')
sns.lineplot(net_rev_reg_month, x = net_rev_reg_month.index, y = net_rev_reg_month.columns[2], marker = 'o', label='South')
sns.lineplot(net_rev_reg_month, x = net_rev_reg_month.index, y = net_rev_reg_month.columns[3], marker = 'o', label='West')
plt.legend()
```

### Ans 9 b)

```python
north_rank = net_rev_reg_month[[('sum', 'discounted_price', 'North')]].rank().dropna()
north_rank.sort_values([('sum', 'discounted_price', 'North')], inplace=True)
best = north_rank.iloc[[0], :]
worst = north_rank.iloc[[-1], :]

print("Best: ", best)
print()
print("Worst: ", worst)
```
