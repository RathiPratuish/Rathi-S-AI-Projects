```python
# Path to the training images folder
training_images_path = "/content/drive/MyDrive/Computer Vision/CV Project/training_images"

# Create a cascade classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

# Loop through all the images in the training images folder
for filename in os.listdir(training_images_path):
    # Read the image
    img = cv2.imread(os.path.join(training_images_path, filename))

    # Convert the image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Detect faces in the image
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    # Draw a rectangle around each detected face
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Display the image with detected faces
    plt.imshow(img)
    plt.show()
```

```python
import os
import cv2
import pandas as pd

# Path to the training images folder
training_images_path = "/content/drive/MyDrive/Computer Vision/CV Project/training_images"

# Create a cascade classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

# Create an empty DataFrame to store the face metadata
face_data = pd.DataFrame()

# Loop through all the images in the training images folder
for filename in os.listdir(training_images_path):
    # Read the image
    img = cv2.imread(os.path.join(training_images_path, filename))

    # Convert the image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Detect faces in the image
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    # Extract metadata for each detected face
    for (x, y, w, h) in faces:
        # Create a dictionary to store the metadata for each face
        face_metadata = {
            "filename": filename,
            "x": x,
            "y": y,
            "width": w,
            "height": h
        }

        # Create a DataFrame from the face_metadata dictionary
        face_metadata_df = pd.DataFrame(face_metadata, index=[0])

        # Concatenate the face_metadata_df DataFrame with the face_data DataFrame
        face_data = pd.concat([face_data, face_metadata_df], ignore_index=True)

# Print the DataFrame with face metadata
print(face_data)
```

The code first defines the path to the training images folder, creates a cascade classifier for face detection, and creates an empty DataFrame to store the face metadata.

```python
import os
import cv2
import pandas as pd

# Path to the training images folder
training_images_path = "/content/drive/MyDrive/Computer Vision/CV Project/training_images"

# Create a cascade classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

# Create an empty DataFrame to store the face metadata
face_data = pd.DataFrame()

# Create an empty DataFrame with a default value for the "total_faces" column
face_data = pd.DataFrame({"total_faces": 0}, index=[0])

# Loop through all the images in the training images folder
for filename in os.listdir(training_images_path):
    # Read the image
    img = cv2.imread(os.path.join(training_images_path, filename))

    # Convert the image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Detect faces in the image
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    # Extract metadata for each detected face
    for (x, y, w, h) in faces:
        # Create a dictionary to store the metadata for each face
        face_metadata = {
            "x": x,
            "y": y,
            "w": w,
            "h": h,
            "total_faces": len(faces),
            "filename": filename
        }

        # Add the face metadata to the DataFrame
        face_data = pd.concat([face_data, pd.DataFrame(face_metadata, index=[0])], ignore_index=True)

# Print the DataFrame with face metadata
print(face_data)
```

```python
# Create an empty DataFrame with the desired column order
face_data = pd.DataFrame(columns=["x", "y", "w", "h", "total_faces", "filename"])

# Loop through all the images in the training images folder
for filename in os.listdir(training_images_path):
    # Read the image
    img = cv2.imread(os.path.join(training_images_path, filename))

    # Convert the image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Detect faces in the image
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    # Extract metadata for each detected face
    for (x, y, w, h) in faces:
        # Create a dictionary to store the metadata for each face
        face_metadata = {
            "x": x,
            "y": y,
            "w": w,
            "h": h,
            "filename": filename
        }

        # Add the face metadata to the DataFrame
        face_data = pd.concat([face_data, pd.DataFrame(face_metadata, index=[0])], ignore_index=True)

    # Add the total number of faces in the image to the DataFrame
    face_data.loc[face_data.shape[0] - len(faces):, "total_faces"] = len(faces)
```

```python
# Print the DataFrame with face metadata
print(face_data)
```
